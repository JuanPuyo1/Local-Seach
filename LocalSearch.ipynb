### a) Funciones de la implementación del método minimax con poda alfa-beta en el juego de “triqui”

Minimax es un método de decisión para minimizar la pérdida máxima esperada en juegos con adversario y con información perfecta. Este cálculo se hace de forma recursiva.

**Método de la solución**

Variables:

* terminal:Se generarán todos los nodos hasta llegar a un estado terminal, cálculo de los valores de la función de utilidad para cada nodo terminal.
* minimax: Calcula el valor de los nodos superiores a partir del valor de los inferiores, se elegirán los valores mínimos y máximos representando los movimientos del jugador y del oponente.
* maximize, minimize: en ambos metodos se hace uso de la poda alfa beta estos hacen que se reduzca el número de nodos a visitar por lo que el tiempo de cálculo sera menor.




```python

class TriquiState:
    def __init__(self, board, move_i=-1, move_j=-1, alpha=None, beta=None):
        if alpha is not None:
            self.alpha = alpha
        else: 
            self.alpha =-2 # -inf
        if beta is not None:
            self.beta = beta
        else: 
            self.beta = 2   # +inf
        
        self.move_i=move_i
        self.move_j=move_j

        self.board=[['_','_','_'],
                    ['_','_','_'],
                    ['_','_','_']]
        if board is not None:
            for i in range(3):
                for j in range(3):
                    self.board[i][j]=board[i][j]
    #El metodo terminal permite evaluar si el estado actual es un estado terminal
    def terminal(self):
        #Evaluar estado terminal por filas
        for i in range(3): 
            if(self.board[i] == ['X','X','X']):
                return 1
            if(self.board[i] == ['O','O','O']):
                return -1
        
        #Evaluar estado terminal por columnas
        for j in range(3):
            X = 0
            O = 0
            for i in range(3):
                if( self.board[i][j] == 'X'):
                    X = X + 1
                if( self.board[i][j] == 'O'):
                    O = O + 1
            if (X == 3):
                return 1
            elif (O == 3):
                return -1
        
        # Evaluar estado terminal, diagonal 1
        if( self.board[0][0] == self.board[1][1] and self.board[1][1] == self.board[2][2] and self.board[0][0] == 'X' ):
            return 1
        if( self.board[0][2] == self.board[1][1] and self.board[1][1] == self.board[2][0] and self.board[0][2] == 'X' ):
            return 1
        
        #Evaluar estado terminal, diagonal 2
        if( self.board[0][0] == self.board[1][1] and self.board[1][1] == self.board[2][2] and self.board[0][0] == 'O' ):
            return -1
        if( self.board[0][2] == self.board[1][1] and self.board[1][1] == self.board[2][0] and self.board[0][2] == 'O' ):
            return -1
        
        # Si el tablero esta vacio, se retorna None para continuar el juego.
        for i in range(3):
            for j in range(3):
                if (self.board[i][j] == '_'):
                    return None
        
        # En otro caso, se trata de un empate
        return 0
    
    def get_board(self):
        another_board=[['_','_','_'],
                       ['_','_','_'],
                       ['_','_','_']]
        for i in range(3):
            for j in range(3):
                another_board[i][j] = self.board[i][j]
        return another_board
    
    def print_board(self):
        print("Utilidad:", self.terminal())
        for i in range(0, 3):
            for j in range(0, 3):
                print('{}|'.format(self.board[i][j]), end=" ")
            print()
        print()
```

```python
# Escribe la solución en pantalla
def print_position(board, n): 
    for i in range(n): 
        for j in range(n): 
            print (board[i][j], end = " ") 
        print() 

# Evalúa la validez de una nueva posición según las damas puestas a la izquierda
def valid_placement(board, row, col, n): 
    #Escribe la posición evaluada y devuelve el tablero a su estado inicial
    board[row][col]=1

    x,y = row,col

    print("Posición evaluada")
    print_position(board,n)
    board[row][col]=0

    # Valida que no haya damas en columnas previas a la actual para la misma fila
    for i in range(col): 
        if board[row][i] == 1:
            print("Hay dama")
            return False
  
    # Valida que no haya damas en la diagonal hacia arriba
    for i in range(row, -1, -1):
        if board[row-i][col-i] == 1:
            print("Hay dama") 
            return False

    # Valida que no haya damas en la diagonal hacia abajo
    for i in range(row, n, 1): 
        if col-i<0 or row+i>n-1: # previene evaluaciones por fuera del tablero
            break
        if board[row+i][col-i] == 1: #Evalua cruces por diagonales
            print("HOLA") 
            return False
        if board[row+i][col-1] == 1:
            print("HOLAer") 
            return False
        if board[row+1][col-1] == 1:
            print("HOLAer") 
            return False

    return True
```

```python
def BTSolve4Queens(board, col, n): 
    # Retorno en el llamado recursivo: todas las damas han sido ubicadas en el tablero
    # (de las columnas 0 a n-1)
    if col == n:

      return True
  
    # Si no se han ubicado todas las damas, ubicar una nueva dama en las casillas
    # de la columna bajo evaluación (col)
    for i in range(n): 
        if valid_placement(board, i, col, n): 
            # Se ubica una nueva dama en la fila i-ésima de esa nueva columna 
            board[i][col] = 1
            x = i
            print(x)
            y = col
            print(y)
            
            # Se hace el llamado recursivo para ubicar el resto de las damas 
            if BTSolve4Queens(board, col + 1, n) == True: 
                return True
  
            # "else", Si no hubo una solución (el valor de retorno fue falso), quitamos la dama evaluada
            board[i][col] = 0

    # Si la dama no pudo ser ubicada en una casilla sobre esta columna para todas las combinaciones
    # previas, se retorna "Falso"
    return False
```
